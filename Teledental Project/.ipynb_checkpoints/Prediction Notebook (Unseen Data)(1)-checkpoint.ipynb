```python
image_size = 224
def downsize(image):
    dsize = (image_size, image_size)
    downsized_image = cv2.resize(image, dsize)
    return downsized_image


def preprocess_image(image):
    image = np.array(image)
    # reshape into shape [batch_size, height, width, num_channels]
    img_reshaped = tf.reshape(image, [1, image.shape[0], image.shape[1], image.shape[2]])
    # Use `convert_image_dtype` to convert to floats in the [0,1] range.
    image = tf.image.convert_image_dtype(img_reshaped, tf.float32)
    return image
```

```python
with open(model_dir + 'imagenet/ImageNetLabels.txt', "r") as f:
    labels = f.readlines()
    classes = [l.strip() for l in labels]
    
def check_human(picture_file):
    image = Image.open(picture_file)
    image = image.resize((image_size, image_size))
    image = preprocess_image(image)

    # Run model on image
    probabilities = tf.nn.softmax(classifier(image)).numpy()
    top_5 = tf.argsort(probabilities, axis=-1, direction="DESCENDING")[0][:8].numpy()
    np_classes = np.array(classes)

    # Some models include an additional 'background' class in the predictions, so
    # we must account for this when reading the class labels.
    includes_background_class = probabilities.shape[1] == 1001

    found_classes = []
    for i, item in enumerate(top_5):
        class_index = item if includes_background_class else item + 1
        # line = f'({i+1}) {class_index:4} - {classes[class_index]}: {probabilities[0][top_5][i]}'
        found_classes.append({classes[class_index]})
    # show_image(image, '')

    # removing {} from found_classes elements
    tmp_list = []
    for i in range(len(found_classes)):
        tmp_list.append(str(found_classes[i])[2:-2])

    found_classes = tmp_list

    # strategy 1 need more classes, I focused on strategy 2
    acceptable_classes = {'lipstick', 'ice lolly', 'neck brace', 'sunscreen', 'mask', 'Band Aid',
                          'face powder'}  # ,'wig'}    #print(found_classes)
    if set(found_classes).intersection(acceptable_classes) != set():  # means there is a class to human face
        acceptance_open_angle = ''
        acceptance_smile_angle = ''
        message = "the photo contains human being"
        human_photo = True
    else:  # images not related to human being
        acceptance_open_angle = False
        acceptance_smile_angle = False
        message = "the photo doesn't contain human being"
        human_photo = False
    print ("check human output: ", human_photo )
    return human_photo  # message, found_classes


def human_classification(uploaded_image):
    image = Image.open(uploaded_image)
    image = image.resize((image_size, image_size))
    image = preprocess_image(image)
    human_photo = check_human(uploaded_image)
    return human_photo
```

```python
#ref: https://github.com/imneonizer/How-to-find-if-an-image-is-bright-or-dark/blob/master/demo.ipynb
'''
LAB Colorspace
L - Lightness
A - Colors from Green to Magenta
B - Colors from Blue to Yellow
'''
def img_estim(image, dim=10):
    image = cv2.resize(image, (dim, dim))
    L, A, B = cv2.split(cv2.cvtColor(image, cv2.COLOR_BGR2LAB)) 
    L = L/np.max(L)
    return np.mean(L) 
```
