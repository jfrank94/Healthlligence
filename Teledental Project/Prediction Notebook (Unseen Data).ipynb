```python
directory = "Validation_Photos"
img_fn = []

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if f[-1] == 'g':
        img_fn.append(f)
```

```python
blur_clf = joblib.load("blur_clf.joblib") 
cb_clf = joblib.load("cb_clf.joblib")
```

```python
#ref: https://github.com/imneonizer/How-to-find-if-an-image-is-bright-or-dark/blob/master/demo.ipynb
'''
LAB Colorspace
L - Lightness
A - Colors from Green to Magenta
B - Colors from Blue to Yellow
'''
def img_estim(image, dim=10):
    image = cv2.resize(image, (dim, dim))
    L, A, B = cv2.split(cv2.cvtColor(image, cv2.COLOR_BGR2LAB)) 
    L = L/np.max(L)
    return np.mean(L) 
```

```python
base_true = []
blur_pred, cb_pred = [], []
target_name = ['True', 'False']

i = 0
for img_f in img_fn:
    img = cv2.imread(img_f)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img) 
    plt.show()
    
    obj = BRISQUE(img_f, url=False)
    brisque_score = obj.score()
    base_true.append(brisque_score)
    print("BRISQUE SCORE: {}".format(brisque_score))
    b_val = blur_clf.predict([[brisque_score]])[0] # output of prediction blurriness
    blur_pred.append(b_val)
    
    light_val = img_estim(img)
    print("Light Value: {}".format(light_val))
    c_val = cb_clf.predict([[light_val]])[0] # output of prediction brightness/contrast
    cb_pred.append(c_val)
    print("[Acceptable Clarity: {}, Acceptable Brightness/Lighting: {}]".format(b_val, c_val))
    print("==="*35)

```
